```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = "C:\\Users\\Gnani\\Desktop\\Laptop Specifications (Responses).xlsx"  
data = pd.read_excel(file_path)

data.drop("Timestamp",axis = 1,inplace = True)
data.info()
```

```python
data = data.fillna(method='ffill')
data.info()
```

```python
plt.figure(figsize=(12,6))
data.groupby('Brand name of your laptop?').size().sort_values(ascending=False).plot(kind = 'bar',color = sns.color_palette('Paired'))
plt.xlabel('Brand')
plt.ylabel('Number of Laptops')
plt.show()
```

```python

plt.figure(figsize=(12,6))
data.groupby('RAM size of your current laptop').size().sort_values(ascending=False).plot(kind = 'bar',color = sns.color_palette('Paired'))
plt.xlabel('Ram Size in GB')
plt.ylabel('Number of Laptops')
plt.show()
```

```python
plt.figure(figsize=(12,6))
data.groupby('Storage type').size().sort_values(ascending=False).plot(kind = 'bar',color = sns.color_palette('Paired'))
plt.xlabel('EStorage ')
plt.ylabel('Number of laptops')
plt.show()
```

```python
plt.figure(figsize=(12,6))
data.groupby('Storage (in GB)').size().sort_values(ascending=False).plot(kind = 'bar',color = sns.color_palette('Paired'))
plt.xlabel('Storage in GB')
plt.ylabel('Number of Laptops')
plt.show()
```

```python
plt.figure(figsize=(12,6))
data.groupby('Processor').size().sort_values(ascending=False).plot(kind = 'bar',color = sns.color_palette('Paired'))
plt.xlabel('Processor type ')
plt.ylabel('Number of Laptops')
plt.show()
```

```python
plt.figure(figsize=(12,6))
data.groupby('Clock speed').size().sort_values(ascending=False).plot(kind = 'bar',color = sns.color_palette('Paired'))
plt.xlabel('Clock speed ')
plt.ylabel('Number of Laptops')
plt.show()
```

```python
plt.figure(figsize=(12,6))
data.groupby('Avg. Battery draining time(hours)').size().sort_values(ascending=False).plot(kind = 'bar',color = sns.color_palette('Paired'))
plt.xlabel('Avg. Battery draining time(hours) ')
plt.ylabel('Number of Laptops')
plt.show()
```

```python
plt.figure(figsize=(12,6))
data.groupby('weight(kgs)').size().sort_values(ascending=False).plot(kind = 'bar',color = sns.color_palette('Paired'))
plt.xlabel('weight(kgs) ')
plt.ylabel('Number of Laptops')
plt.show()
```

```python
plt.figure(figsize=(12,6))
data.groupby('screen size').size().sort_values(ascending=False).plot(kind = 'bar',color = sns.color_palette('Paired'))
plt.xlabel('screen size in inches ')
plt.ylabel('Number of Laptops')
plt.show()
```

```python
plt.figure(figsize=(12,6))
data.groupby('Ethernet port').size().sort_values(ascending=False).plot(kind = 'bar',color = sns.color_palette('Paired'))
plt.xlabel('Ethernet port ')
plt.ylabel('Number of Laptops')
plt.show()
```
